```
#Spark streaming with RDD
from pyspark.streaming import StreamingContext
from pyspark import SparkConf, SparkContext

#instantiating a spark context for RDD
conf = SparkConf().setAppName("Spark_streaming")
sc = SparkContext.getOrCreate(conf = conf)

#Instantiating a Spark Stream context
ssc = StreamingContext(sc,1)
```

# Spark Streaming RDD

```
rdd = ssc.textFileStream("/FileStore/tables/streams/")
```

```
rdd.pprint()
ssc.start()
ssc.awaitTerminationOrTimeout(120)
```

# Spark Streaming Dataframe

```
from pyspark.sql import SparkSession

spark = SparkSession.builder.appName("Streaming DF").getOrCreate()

words = spark.readStream.text("/FileStore/tables/streams")
words_count = words.groupBy("value").count()
display(words_count)
```
