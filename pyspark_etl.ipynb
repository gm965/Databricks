```
from pyspark.sql import SparkSession
spark = SparkSession.builder.appName("ETL").getOrCreate()
```

```
#Extract the data
df = spark.read.text("/FileStore/tables/ETL/generic_statement_for_ETL.txt")
display(df)
```

```
#Transform the data
from pyspark.sql.functions import lit, col, explode
import pyspark.sql.functions as f
```

```
df1 = df.withColumn("splitcol", f.split("value"," "))
df2 = df1.withColumn("word",explode("splitcol"))
words_df = df2.select("word")
word_cnt = words_df.groupBy("word").count()
word_cnt.show()
```

```
#Loading data in RDS
driver = "org.postgresql.Driver"
url = "jdbc:postgresql://database-1.cjhi5plc09le.ap-south-1.rds.amazonaws.com/"
table = "RDS_pyspark.wordcount"
username = "postgres"
password = "postgres123"

word_cnt.write.format("jdbc").option("driver",driver).option("url",url).option("dbtable", table).option("mode","append").option("user",username).option("password",password).save()
```
